```python
from pylab import *
import numpy as np
def trends(start,end,prop,fit):
    data_range=data.loc[(data["Date"]>=start)& (data["Date"]<=end)] #select data from a range
    data_range["Date"]=pd.to_datetime(data_range["Date"], format='%Y-%m-%d')
    data_range.set_index(["Date"],inplace=True)
    x=arange(len(data_range)) 
    y=data_range[prop]    
    figure(figsize=(27,8)) # setting fig size
    plot(y)
    z = np.polyfit(x, y, fit)
    p = np.poly1d(z)
    plt.title(prop+" price")
    xlabel("Times")
    ylabel("Price(USD)")
    plot(x, p(x),'r-.')
    show()
```

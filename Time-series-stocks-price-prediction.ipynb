```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Load the time series data
data = pd.read_csv('C:/Users/hieu7/Downloads/AMZN.csv')
```

```python
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
def plots(start,end,prop):
    data_range=data.loc[(data["Date"]>=start)& (data["Date"]<=end)] #select data from a range date=data_range["Date"]
    date=data_range["Date"]
    Open_data=data_range["Open"]
    x = [dt.datetime.strptime(d,'%Y-%m-%d').date() for d in date]
    y=Open_data
    figure(figsize=(27,8)) # setting fig size
    plot(x,y)
    plt.gcf().autofmt_xdate()
    plt.title(prop+" price")
    xlabel("Times")
    ylabel("Price(USD)")
    show()
```

```python
from pylab import *
import numpy as np
def trends(start,end,prop,fit):
    data_range=data.loc[(data["Date"]>=start)& (data["Date"]<=end)] #select data from a range
    x=arange(len(data_range)) 
    y=data_range[prop] 
    figure(figsize=(27,8)) # setting fig size
    scatter(x,y)
    z = np.polyfit(x, y, fit)
    p = np.poly1d(z)
    plt.title(prop+" price")
    xlabel("Times")
    ylabel("Price(USD)")
    plot(x, p(x),'r-.')
    show()
```

```python
start="2021-03-27"
end="2023-03-27"
prop="Open"
fit=2
trends(start,end,prop,fit)
```

```python
plots(start,end,prop)
```
